```python
import matplotlib as plt
import numpy as np
from sklearn import linear_model, datasets
```

```python
diabetes= datasets.load_diabetes()
```

```python
#diabetes
```

```python
print(diabetes.keys())
```

```python
print(diabetes.DESCR)
```

```python
print(diabetes.data)
```

```python
print(diabetes.feature_names)
```

```python
diabetes_X= diabetes.data[:,np.newaxis,2]
```

```python
diabetes_X
```

### Train_Test_ Split

```python
diabetes_X_train= diabetes_X[:-30]
```

```python
diabetes_X_test= diabetes_X[-30:]
```

```python
diabetes_X_train                                # Training Data 70%
```

```python
diabetes_X_test                                     # Testing Data 30%
```

```python
diabetes_Y_train= diabetes.target[:-30]
```

```python
diabetes_Y_train
```

```python
diabetes_Y_test= diabetes.target[-30:]
```

```python
diabetes_Y_test
```

### LINEAR MODEL

```python
model= linear_model.LinearRegression()
```

```python
model.fit(diabetes_X_train, diabetes_Y_train)
```

```python
diabetes_Y_predicted= model.predict(diabetes_X_test)
diabetes_Y_predicted
```

### Predictions

```python
diabetes_Y_predicted= model.predict([[0.01876761]])
diabetes_Y_predicted
```

### Accuracy: MSE(Mean Squred Error)

```python
from sklearn.metrics import mean_squared_error
```

```python
print("Mean Squared Error is: ",mean_squared_error(diabetes_Y_test,diabetes_Y_predicted))
```

```python
print("Weights: ",model.coef_,"\nIntercept: ",model.intercept_)
```
