```python
from sklearn import datasets
```

```python
iris= datasets.load_iris()
```

```python
print(iris.DESCR)
```

```python
print(list(iris.keys()))
```

```python
print(iris.data)
```

```python
print(iris.target)
```

```python
iris.data.shape
```

```python
X= iris.data[:,3:]
```

```python
X
```

```python
#X= iris["data"][:,3:]
```

```python
Y= (iris["target"]==2)
```

```python
Y
```

```python
import numpy as np
```

```python
Y= (iris["target"]==2).astype(np.int)
```

```python
Y
```

### Train Model Logistic Regression Classifier --- X, Y

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf= LogisticRegression()
```

```python
clf.fit(X,Y)
```

```python
example= clf.predict([[1.6]])
```

```python
example                                     #not iris verginica
```

```python
example= clf.predict([[2.5]])
```

```python
example                                     #iris verginica
```

### Data Visualization

```python
X_new= np.linspace(0,3,1000).reshape(-1,1)
```

```python
X_new
```

```python
Y_prob= clf.predict_proba(X_new)
```

```python
Y_prob
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X_new,Y_prob[:,1],"g-", label="virginica")
```
